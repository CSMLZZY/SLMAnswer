```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Percp:
    def __init__(self):
        self.w = np.zeros(X.shape[1])
        self.b = 0
        self.rate = 1

    def term_sum(self, X, w, b):
        return np.dot(X, w) + b

    def fit(self, X_train, y_train):
        notEnd = True
        errSum = 0
        while notEnd:
            errSum = 0
            for i in range(X_train.shape[0]):
                if ((y_train[i] * self.term_sum(X_train[i], self.w, self.b)) <= 0):
                    self.w = self.w + self.rate * y_train[i] * X_train[i]
                    self.b = self.b + self.rate * y_train[i]
                    errSum += 1
            if errSum == 0:
                notEnd = False
        return 'Success!'
```

```python
    X = np.array([[3, 3], [4, 3], [1, 1]])
    y = np.array([1, 1, -1])
    plt.figure(figsize=(12,6))
    plt.subplot(121)
    plt.xlim(0, 6)
    plt.ylim(0, 6)
    plt.scatter(X[:2, :1], X[:2, 1:2], s=50, c='b',
                marker='o', label='Positive')
    plt.scatter(X[-1, :1], X[-1, 1:2], s=50, c='r',
                marker='x', label='Negative')
    plt.legend()
    gzj = Percp()
    gzj.fit(X, y)
    print(gzj.w)
    print(gzj.b)
    plt.subplot(122)
    plt.xlim(0, 6)
    plt.ylim(0, 6)
    plt.scatter(X[:2, :1], X[:2, 1:2], s=50, c='b',
                marker='o', label='Positive')
    plt.scatter(X[-1, :1], X[-1, 1:2], s=50, c='r',
                marker='x', label='Negative')
    X_points = np.linspace(0, 7, 10)
    y_points = -(gzj.w[0] * X_points + gzj.b) / gzj.w[1]
    plt.plot(X_points, y_points)
    plt.legend()
    plt.show()
```
